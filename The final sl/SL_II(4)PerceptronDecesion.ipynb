```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset (linearly separable)
np.random.seed(1)
X = np.random.randn(100, 2)
y = np.where(X[:, 0] + X[:, 1] > 0, 1, -1)  # Linearly separable classes

# Initialize weights and bias
weights = np.random.rand(2)
bias = np.random.rand()
learning_rate = 0.1
epochs = 10

# Perceptron Learning Algorithm
for epoch in range(epochs):
    for i in range(len(X)):
        x_i = X[i]
        y_pred = np.sign(np.dot(x_i, weights) + bias)
        error = y[i] - y_pred
        # Update rule
        weights += learning_rate * error * x_i
        bias += learning_rate * error

# Decision Boundary Visualization
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = np.sign(weights[0] * xx + weights[1] * yy + bias)

# Plot the decision regions
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors="k", marker="o")
plt.title("Perceptron Learning - Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
