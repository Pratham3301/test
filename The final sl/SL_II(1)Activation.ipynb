```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
```

**Activation Functions**

```python
# Linear Activation Function

def linear(x):
    return x

# Generate input values
x = np.linspace(-6, 6, 600)
y = linear(x)

# Plot
plt.figure(figsize=(7,5))
plt.plot(x, y, label="Linear Activation", color="blue")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Linear Activation Function")
plt.legend()
plt.grid()
plt.show()
```

```python
# Sigmoid Activation Function

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y = sigmoid(x)

plt.figure(figsize=(7,5))
plt.plot(x, y, label="Sigmoid Activation", color="green")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Sigmoid Activation Function")
plt.legend()
plt.grid()
plt.show()
```

```python
# ReLU Activation Function

def relu(x):
    return np.maximum(0, x)

y = relu(x)

plt.figure(figsize=(7,5))
plt.plot(x, y, label="ReLU Activation", color="red")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("ReLU Activation Function")
plt.legend()
plt.grid()
plt.show()
```

```python
# Tanh Activation Function

def tanh(x):
    return np.tanh(x)

y = tanh(x)

plt.figure(figsize=(7,5))
plt.plot(x, y, label="Tanh Activation", color="purple")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Tanh Activation Function")
plt.legend()
plt.grid()
plt.show()
```

```python
# Leaky ReLU Activation Function

def leaky_relu(x, alpha=0.03):
    return np.where(x > 0, x, alpha * x)

y = leaky_relu(x)

plt.figure(figsize=(7,5))
plt.plot(x, y, label="Leaky ReLU Activation", color="orange")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Leaky ReLU Activation Function")
plt.legend()
plt.grid()
plt.show()
```

```python
# Softmax Activation Function

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Stability trick
    return exp_x / np.sum(exp_x)

y = softmax(x)

plt.figure(figsize=(7,5))
plt.plot(x, y, label="Softmax Activation", color="brown")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Softmax Activation Function")
plt.legend()
plt.grid()
plt.show()
```

```python
# ELU Activation Function

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

y = elu(x)

plt.figure(figsize=(7,5))
plt.plot(x, y, label="ELU Activation", color="pink")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("ELU Activation Function")
plt.legend()
plt.grid()
plt.show()
```

```python
# Softplus Activation Function

def softplus(x):
    return np.log1p(np.exp(x))  # log(1+e^x) is more stable

y = softplus(x)

plt.figure(figsize=(7,5))
plt.plot(x, y, label="Softplus Activation", color="cyan")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Softplus Activation Function")
plt.legend()
plt.grid()
plt.show()
```

```python
# SiLU (Swish) Activation Function

def silu(x):
    return x * sigmoid(x)  # Swish Function

y = silu(x)

plt.figure(figsize=(7,5))
plt.plot(x, y, label="SiLU (Swish) Activation", color="black")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("SiLU (Swish) Activation Function")
plt.legend()
plt.grid()
plt.show()
```
